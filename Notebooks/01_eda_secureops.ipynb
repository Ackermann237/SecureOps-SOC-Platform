# Construire le score + le niveau de risque 

```python
import numpy as np
import pandas as pd

# âœ… Assure-toi que ces colonnes existent dÃ©jÃ  (tu les as crÃ©Ã©es)
# high_failed_logins, low_ip_reputation, unusual_access, no_encryption

WEIGHTS = {
    "low_ip_reputation": 40,
    "unusual_access": 30,
    "high_failed_logins": 20,
    "no_encryption": 10
}

# (Optionnel) si attack_detected n'est pas propre, sÃ©curise-le:
# intrusion_df["attack_detected"] = intrusion_df["attack_detected"].astype(int)

# âœ… Score sur 100
intrusion_df["risk_score"] = (
    intrusion_df["low_ip_reputation"].astype(int) * WEIGHTS["low_ip_reputation"]
    + intrusion_df["unusual_access"].astype(int) * WEIGHTS["unusual_access"]
    + intrusion_df["high_failed_logins"].astype(int) * WEIGHTS["high_failed_logins"]
    + intrusion_df["no_encryption"].astype(int) * WEIGHTS["no_encryption"]
)

# âœ… CatÃ©gories de risque
def risk_level(score):
    if score >= 70:
        return "High"
    elif score >= 40:
        return "Medium"
    else:
        return "Low"

intrusion_df["risk_level"] = intrusion_df["risk_score"].apply(risk_level)

# âœ… Explication : quelles rÃ¨gles ont dÃ©clenchÃ© le score (lisible SOC)
def explain_row(row):
    triggers = []
    if row["low_ip_reputation"]:
        triggers.append("Low IP reputation")
    if row["unusual_access"]:
        triggers.append("Unusual access time")
    if row["high_failed_logins"]:
        triggers.append("High failed logins")
    if row["no_encryption"]:
        triggers.append("No encryption")
    return ", ".join(triggers) if triggers else "No strong indicator"

intrusion_df["risk_drivers"] = intrusion_df.apply(explain_row, axis=1)

intrusion_df[["session_id","risk_score","risk_level","risk_drivers","attack_detected"]].head(10)
```

Mise en place dâ€™un modÃ¨le de dÃ©tection dâ€™anomalies rÃ©seau (Isolation Forest) sans supervision, intÃ©grÃ© Ã  un moteur de rÃ¨gles SOC explicable pour la dÃ©tection de comportements suspects en temps rÃ©el.

```python
# ==========================================================
# Ã‰TAPE 4 â€” CHARGEMENT CONSOLIDÃ‰ (EXPORT DATASETS PROCESSED)
# Objectif : produire un dataset final consolidated_soc.csv
# ==========================================================

from pathlib import Path
import pandas as pd
import numpy as np

# -----------------------------
# 1) DÃ©finir les dossiers de sortie
# -----------------------------
BASE_DIR = Path(".")  # si ton notebook est Ã  la racine du projet
PROCESSED_DIR = BASE_DIR / "data" / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

# -----------------------------
# 2) SECURISER / COMPLETER LES FEATURES SI BESOIN
# -----------------------------

# ========== A) Intrusion (Logs rÃ©seau) ==========
intrusion_out = intrusion_df.copy()

# Normaliser la date si elle n'existe pas (dataset intrusion n'a pas forcÃ©ment timestamp)
# => On crÃ©e une "event_date" fictive basÃ©e sur l'index (utile pour KPI journaliers)
if "event_date" not in intrusion_out.columns:
    intrusion_out["event_date"] = pd.Timestamp.today().normalize()

# S'assurer que high_risk_session existe
if "high_risk_session" not in intrusion_out.columns:
    intrusion_out["high_risk_session"] = (
        (intrusion_out.get("failed_logins_count", 0) >= 3) |
        (intrusion_out.get("ip_reputation_score", 1) < 0.3) |
        (intrusion_out.get("unusual_time_access", 0) == 1)
    )

# Si anomaly n'existe pas, on met 0 (au cas oÃ¹ tu nâ€™as pas encore exÃ©cutÃ© IsolationForest)
if "anomaly" not in intrusion_out.columns:
    intrusion_out["anomaly"] = 0

# ========== B) Incidents (SOC incidents) ==========
incidents_out = incidents_df.copy()

# date_reported est ta colonne principale (tu lâ€™as dÃ©jÃ )
if "date_reported" in incidents_out.columns:
    incidents_out["event_date"] = pd.to_datetime(incidents_out["date_reported"], errors="coerce").dt.date
else:
    # fallback
    incidents_out["event_date"] = pd.Timestamp.today().normalize().date()

# Calcul durÃ©e incident si pas dÃ©jÃ  prÃ©sent
if "incident_duration_days" not in incidents_out.columns:
    if ("breach_start" in incidents_out.columns) and ("breach_end" in incidents_out.columns):
        incidents_out["incident_duration_days"] = (
            (pd.to_datetime(incidents_out["breach_end"], errors="coerce") -
             pd.to_datetime(incidents_out["breach_start"], errors="coerce"))
            .dt.days
        )
    else:
        incidents_out["incident_duration_days"] = np.nan

# Si critical_incident n'existe pas, on le dÃ©rive (ex: threshold sur individuals_affected)
if "critical_incident" not in incidents_out.columns:
    if "individuals_affected" in incidents_out.columns:
        incidents_out["critical_incident"] = (incidents_out["individuals_affected"] >= 10000).astype(int)
    else:
        incidents_out["critical_incident"] = 0

# ========== C) Tickets IT (ITSM) ==========
tickets_out = tickets_df.copy()

# CrÃ©ation d'une date d'Ã©vÃ¨nement basÃ©e sur Created time
if "Created time" in tickets_out.columns:
    tickets_out["event_date"] = pd.to_datetime(tickets_out["Created time"], errors="coerce").dt.date
else:
    tickets_out["event_date"] = pd.Timestamp.today().normalize().date()

# Ton champ resolution_time_minutes tu lâ€™as dÃ©jÃ  crÃ©Ã©, mais on sÃ©curise
if "resolution_time_minutes" not in tickets_out.columns:
    if "Resolution time" in tickets_out.columns:
        tickets_out["resolution_time_minutes"] = (
            pd.to_timedelta(tickets_out["Resolution time"], errors="coerce")
            .dt.total_seconds() / 60
        )
    else:
        tickets_out["resolution_time_minutes"] = np.nan

# -----------------------------
# 3) EXPORT DES 3 TABLES PROCESSED
# -----------------------------
intrusion_path = PROCESSED_DIR / "intrusion_processed.csv"
incidents_path = PROCESSED_DIR / "incidents_processed.csv"
tickets_path   = PROCESSED_DIR / "tickets_processed.csv"

intrusion_out.to_csv(intrusion_path, index=False)
incidents_out.to_csv(incidents_path, index=False)
tickets_out.to_csv(tickets_path, index=False)

print("âœ… Export OK :")
print(" -", intrusion_path)
print(" -", incidents_path)
print(" -", tickets_path)

# -----------------------------
# 4) CONSTRUCTION DU DATASET CONSOLIDÃ‰ (KPI JOURNALIERS)
# -----------------------------

# A) KPI Intrusion (logs)
intrusion_daily = intrusion_out.groupby(pd.to_datetime(intrusion_out["event_date"]).dt.date).agg(
    total_sessions=("session_id", "count") if "session_id" in intrusion_out.columns else ("anomaly", "count"),
    anomalies_detected=("anomaly", "sum"),
    high_risk_sessions=("high_risk_session", "sum"),
    avg_packet_size=("packet_size", "mean") if "packet_size" in intrusion_out.columns else ("anomaly", "mean"),
    avg_ip_reputation=("ip_reputation_score", "mean") if "ip_reputation_score" in intrusion_out.columns else ("anomaly", "mean")
).reset_index().rename(columns={"event_date": "date"})

# B) KPI Incidents
incidents_daily = incidents_out.groupby(pd.to_datetime(incidents_out["event_date"]).dt.date).agg(
    total_incidents=("incident_id", "count") if "incident_id" in incidents_out.columns else ("critical_incident", "count"),
    critical_incidents=("critical_incident", "sum"),
    avg_incident_duration_days=("incident_duration_days", "mean")
).reset_index().rename(columns={"event_date": "date"})

incidents_daily["critical_incident_rate"] = (
    incidents_daily["critical_incidents"] / incidents_daily["total_incidents"]
).replace([np.inf, -np.inf], np.nan)

# C) KPI Tickets
tickets_daily = tickets_out.groupby(pd.to_datetime(tickets_out["event_date"]).dt.date).agg(
    total_tickets=("Ticket ID", "count") if "Ticket ID" in tickets_out.columns else ("Status", "count"),
    avg_resolution_minutes=("resolution_time_minutes", "mean"),
    p95_resolution_minutes=("resolution_time_minutes", lambda x: np.nanpercentile(x.dropna(), 95) if x.dropna().shape[0] > 0 else np.nan)
).reset_index().rename(columns={"event_date": "date"})

# -----------------------------
# 5) MERGE FINAL (CONSOLIDATED_SOC)
# -----------------------------
consolidated = (
    intrusion_daily.merge(incidents_daily, on="date", how="outer")
                  .merge(tickets_daily, on="date", how="outer")
)

# Remplir NaN KPI par 0 lÃ  oÃ¹ Ã§a a du sens
fill_zero_cols = [
    "total_sessions", "anomalies_detected", "high_risk_sessions",
    "total_incidents", "critical_incidents", "total_tickets"
]
for c in fill_zero_cols:
    if c in consolidated.columns:
        consolidated[c] = consolidated[c].fillna(0).astype(int)

# Trier par date
consolidated["date"] = pd.to_datetime(consolidated["date"])
consolidated = consolidated.sort_values("date")

# Export final
consolidated_path = PROCESSED_DIR / "consolidated_soc.csv"
consolidated.to_csv(consolidated_path, index=False)

print("\nâœ… CONSOLIDATION OK :")
print(" -", consolidated_path)
display(consolidated.head(10))
```
